```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv(r'Datasets\raw.githubusercontent.com_flaviovdf_evcomp2018_master_data_educ_inc.csv')


filtered_data = data[(data['year'] == 2014) & (data['age'] > 17)]


education_income_data = filtered_data[['Educational Attainment', 'Personal Income']]

# Step 4: Further Analysis
# Summary statistics
summary_stats = education_income_data.groupby('Educational Attainment')['Personal Income'].describe()

# Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(data=education_income_data, x='Educational Attainment', y='Personal Income')
plt.xticks(rotation=45)
plt.xlabel('Educational Attainment')
plt.ylabel('Personal Income')
plt.title('Income Distribution by Educational Attainment (2014)')
plt.tight_layout()
plt.show()
```
